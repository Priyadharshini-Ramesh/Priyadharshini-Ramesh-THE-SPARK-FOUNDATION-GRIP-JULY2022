**Upload the dataset**



```python
import pandas as pd
import numpy as np
data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
data.shape
```

```python
data.columns
```

```python
data
```

**Split the data for train and test**




```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(data,train_size=0.8,test_size=0.2)
```

```python
train.shape
```

```python
test.shape
```

```python
train_x=train.drop("Scores",axis=1)
train_y=train["Scores"]
```

```python
train_x.shape
```

```python
train_y.shape
```

```python
test_x=test.drop("Scores",axis=1)
test_y=test["Scores"]
```

```python
test_x.shape
```

```python
test_y.shape
```

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(train_x, train_y)
```

```python
import matplotlib.pyplot as plt
```

```python
reg.coef_ 
```

```python
reg.intercept_ 
```

**Formula for Straight Line** = aX + b

```python
st_line = train_x * reg.coef_ + reg.intercept_
plt.scatter(train_x,train_y)
plt.plot(train_x,st_line)
plt.title('Test Data')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

**Prediction**

```python
Prediction_y = reg.predict(test_x)
P = zip(test_y,Prediction_y)
list(P)
```

```python
df = pd.DataFrame({'Actual': test_y, 'Predicted': Prediction_y})  
df 
```

**MEAN SQUARED ERROR**

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
mse(test_y,Prediction_y,squared=False)
```
