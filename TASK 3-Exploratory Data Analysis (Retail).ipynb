# MULTIVARIATE ANALYSIS



```python
data.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['green','lightblue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

# **Observation 3:**

In Category, maximum profit in Technology and loss in Furniture.

```python
data.groupby('Sub-Category')[['Profit','Sales']].sum().plot.bar(color=['green','red'],figsize=(20,10))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

# **Observation 4:**

In Sub-Category, Copiers attain the maximum profit and loss in Tables.

```python
data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['darkblue','skyblue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

# **Observation 5:**

The West and East region has highest Sales and Profit and loss in central or south.

```python
data.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['green','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

# **Observation 6:**

The maximum profit and sales in the segement of Consumer.

```python
data.groupby('Ship Mode')[['Profit','Sales']].sum().plot.bar(color=['green','lightblue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

# **Observation 7:**

The maximum profit in Standard Class and loss in Same Day.
