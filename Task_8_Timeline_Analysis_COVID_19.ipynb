# ***Highest 10 Deaths in  countries wise:***

```python
hd = data[data['Deaths'] > 0][['Country/Region', 'Deaths']].sort_values(by = 'Deaths', ascending = False).head(10).set_index('Country/Region').plot(kind = 'bar', figsize = (12,8))
plt.title('Highest number of deaths in the World', size=20, color = 'darkblue')
plt.xlabel('Country', size=15, color='r')
plt.ylabel('No., of Deaths', size=15, color='r')
plt.xticks(size=15)
plt.yticks(size=15)
```

# ***Highest confirmed cases:***

```python
hr = data[data['Confirmed'] > 0][['Country/Region','Confirmed', 'Deaths', 'Recovered']].head().sort_values(by = 'Confirmed', ascending = False).set_index('Country/Region').plot(kind = 'line', figsize = (18,8))
plt.title('Highest number of confirmed cases in the world', size=20, color = 'darkblue')
plt.xlabel('Country', size=15, color='r')
plt.ylabel('No., of Cases', size=15, color='r')
plt.xticks(size=15)
plt.yticks(size=15)
```

# ***Highest Death compare to Recovery cases:***

```python
data[data['Recovered'] < data['Deaths']][['Country/Region', 'Active', 'Confirmed', 'Recovered', 'Deaths']].set_index('Country/Region').plot(kind = 'line', figsize = (18,6))
plt.title('Highest number of deaths cases compared to recovery cases', size=20, color = 'darkblue')
plt.xlabel('Country', size=15, color='r')
plt.ylabel('No., of Cases', size=15, color='r')
plt.xticks(size=15)
plt.yticks(size=15)
```

# ***Growth of the cases in the Region:***

```python
data.groupby('WHO Region')[['Confirmed', 'Deaths']].max().sort_values(by = 'Confirmed', ascending = False).plot(kind = 'line', figsize = (16,6))
plt.title('Growth of the cases in the world', size=20, color = 'darkblue')
plt.xlabel('Country', size=15, color='r')
plt.ylabel('No., of Cases', size=15, color='r')
plt.xticks(size=15)
plt.yticks(size=15)
```

# ***Growth % by a week in the world:***

```python
data.groupby(['WHO Region'])['1 week % increase'].sum().plot(kind = 'bar', figsize = (10,5))
plt.title('Growth % by a week in the world', size=20, color = 'darkblue')
plt.xlabel('Region', size=15, color='r')
plt.ylabel('No., of %', size=15, color='r')
plt.xticks(size=15)
plt.yticks(size=15)
```

# ***Correlation:***

```python
plt.figure(figsize = (18,10))
sb.heatmap(data.corr(),annot = True)
```

```python
plt.figure(figsize=(18,10))
sb.lineplot(x='New cases',y='New deaths',data=data)
plt.title('Total deaths across countries',size=20)
plt.ylabel('New deaths',size=15)
plt.xlabel('New cases',size=15)
plt.show()
```
